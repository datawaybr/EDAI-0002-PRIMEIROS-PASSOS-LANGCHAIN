```python
from langchain_chroma import Chroma

page_url = "https://python.langchain.com/docs/how_to/vectorstore_retriever/"

loader = WebBaseLoader(web_paths=[page_url])
docs = loader.load()

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=300,
    chunk_overlap=20,
    length_function=len,
)
splited_docs = text_splitter.split_documents(docs)
db = Chroma.from_documents(splited_docs, OpenAIEmbeddings())
```

```python
db.similarity_search("Como eu posso criar um retreiever a partir de uma vectorstore?")
```

```python
db.similarity_search("O que √© o top_k?")
```

```python
db.similarity_search_with_score("Para que serve o parametro top_k")
```

```python
db.similarity_search_with_relevance_scores("Para que serve o parametro top_k")
```

```python
retriever = db.as_retriever(
    search_type="similarity_score_threshold", search_kwargs={"score_threshold": 0.6}
)
```

```python
retriever.invoke("Para que serve o parametro top_k")
```

```python
from langchain import hub
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser
from langchain_openai import ChatOpenAI

prompt = hub.pull("rlm/rag-prompt")
llm = ChatOpenAI(model="gpt-4o-mini")
```

```python
prompt.messages[0]
```

```python
RAG_chain = (
    {"context": retriever, "question": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)
```

```python
RAG_chain.invoke("Para que serve o parametro top_k")
```

```python
RAG_chain_sem_parser = (
    {"context": retriever, "question": RunnablePassthrough()}
    | prompt
    | llm
)
```

```python
response = RAG_chain_sem_parser.invoke("Para que serve o parametro top_k")
response
```

```python
from langchain_core.runnables import RunnableParallel

def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)

rag_chain_funcao = (
    RunnablePassthrough.assign(context=(lambda x: format_docs(x["context"])))
    | prompt
    | llm
    | StrOutputParser()
)

rag_chain_fontes = RunnableParallel(
    {"context": retriever, "question": RunnablePassthrough()}
).assign(answer=rag_chain_funcao)

rag_chain_fontes.invoke("Para que serve o parametro top_k")
```
